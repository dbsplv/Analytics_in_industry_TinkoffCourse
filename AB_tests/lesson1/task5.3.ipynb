```python
import numpy as np
import pandas as pd
import scipy.stats as st
import matplotlib.pyplot as plt


from statsmodels.stats.weightstats import _zconfint_generic, _tconfint_generic
from statsmodels.stats.proportion import proportion_confint

```

Имеем, что $\hat{p_a} = \frac{1599}{16500}$ и $\hat{p_b} = \frac{1580}{15903}$  
$H_0: p_a=p_b$  
$H_1: p_b > p_a$  
$$\frac{\hat{p_b} - \hat{p_a}} { \sqrt{p(1-p) (\frac{1}{n_b} + \frac{1}{n_a}  )}}  \sim N(0,1),$$ где  
$$ p = \frac{\hat{p_b}n_b + \hat{p_a}n_a}{n_b+n_a}$$


```python
n_b=15903
n_a=16500
alpha=0.05
p_b_sample=1580/n_b
p_a_sample=1599/n_a

p=(p_b_sample*n_b  +p_a_sample*n_a )/(n_b+n_a)

z_stat = (p_b_sample - p_a_sample)/np.sqrt((p*(1-p)*(1/n_b + 1/n_a) ))
print('Z-статистика = ',z_stat)
print('Критическое значение на уровне значимости {} равно'.format(alpha),  st.norm.ppf(0.95))


if(z_stat< st.norm.ppf(0.95)):
    print('Гипотеза H0 не отвергается, прирост конверсии не статзначимый')
else:
    print('Гипотеза H0 отвергается, прирост конверсии статзначимый')
```
