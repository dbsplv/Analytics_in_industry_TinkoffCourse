# Задача 2

Имеем, что нужно проверить гипотезу о доле, причем $\hat{p} = 0.09$  
$H_0: p = 0.11$  
$H_1: p < 0.11$  
Расчитаем Z-статистику
$$\frac{\hat{p}-p}{\sqrt{\frac{p * (1-p)}{n}}}  \sim N(0,1)$$


```python
p_sample = 0.09
n = 1400
p_test=0.11
```

```python
z_stat= (p_sample - p_test)/(np.sqrt(p_test*(1-p_test) /n ))
z_stat
```

```python
p_value = st.norm.cdf(z_stat)
print('p_value = ',p_value)
```
