```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import math
```

```python
def stat(dist='normal', params=[0,1], N=1000):
    if dist == 'normal':
        y = st.norm.rvs(loc = params[0], scale = params[1], size = N)
        x= np.linspace(st.norm.ppf(0.01,loc = params[0], scale = params[1] ),
                      st.norm.ppf(0.99,loc = params[0], scale = params[1] ),
                      num=1000
                      )
        f_x=st.norm.pdf(x,loc = params[0], scale = params[1])
        
       
        fig, ax = plt.subplots()
        ax.plot(x,f_x)
        ax.set_xlabel('x')
        ax.set_ylabel('f(x)')
        ax.set_title('Функция плотности N({},{})'.format(params[0],params[1]))
        
        
        fig, ax = plt.subplots()
        
        ax.hist(y,density=True)
        ax.set_title('Гистограмма выборки из N({},{})'.format(params[0],params[1]))
        
        print('E_x = ', params[0])
        print('mean_x = ', np.mean(y))
        print('E_x - mean_x = ', params[0] - np.mean(y))
        
    if dist == 'uniform':
        y = st.uniform.rvs(loc = params[0], scale = params[1], size = N)
        x= np.linspace(st.uniform.ppf(0.01,loc = params[0], scale = params[1] ),
                      st.uniform.ppf(0.99,loc = params[0], scale = params[1] ),
                      num=1000
                      )
        f_x=st.uniform.pdf(x,loc = params[0], scale = params[1])
        
       
        fig, ax = plt.subplots()
        ax.plot(x,f_x)
        ax.set_xlabel('x')
        ax.set_ylabel('f(x)')
        ax.set_title('Функция плотности U({},{})'.format(params[0],params[1]))
        
        
        fig, ax = plt.subplots()
        
        ax.hist(y,density=True)
        ax.set_title('Гистограмма выборки из U({},{})'.format(params[0],params[1]))
        
        print('E_x = ', st.uniform.mean(loc = params[0], scale = params[1]))
        print('mean_x = ', np.mean(y))
        print('E_x - mean_x = ', st.uniform.mean(loc = params[0], scale = params[1]) - np.mean(y))
     
    if dist == 'expon':
       
        y = st.expon.rvs(scale = 1/params[0], size = N)
        x= np.linspace(st.expon.ppf(0.01,scale = 1/params[0] ),
                      st.expon.ppf(0.99,scale = 1/params[0] ),
                      num=1000
                      )
        f_x=st.expon.pdf(x,scale = 1/params[0])
        
       
        fig, ax = plt.subplots()
        ax.plot(x,f_x)
        ax.set_xlabel('x')
        ax.set_ylabel('f(x)')
        ax.set_title('Функция плотности exp({})'.format(params[0]))
        
        
        fig, ax = plt.subplots()
        
        ax.hist(y,density=True)
        ax.set_title('Гистограмма выборки из exp({})'.format(params[0]))
        
        print('E_x = ', st.expon.mean(scale = 1/params[0]))
        print('mean_x = ', np.mean(y))
        print('E_x - mean_x = ', st.expon.mean(scale = 1/params[0]) - np.mean(y))
    if dist == 'binom':
        n=params[0]
        p=params[1]
        y = st.binom.rvs(n,p, size = N)
        x= np.arange(st.binom.ppf(0.01,n,p ),
                      st.binom.ppf(0.99,n,p )
                
                      )
        
        fig, ax = plt.subplots()
        ax.plot(x, st.binom.pmf(x, n, p), 'bo', ms=8, label='binom pmf')
    
        ax.set_xlabel('x')
        ax.set_ylabel('f(x)')
        ax.set_title('Функция плотности Bin({},{})'.format(params[0],params[1]))
        
        
        fig, ax = plt.subplots()
        
        ax.hist(y,density=True)
        ax.set_title('Гистограмма выборки из Bin({},{})'.format(params[0],params[1]))
        
        print('E_x = ', st.binom.mean(n,p))
        print('mean_x = ', np.mean(y))
        print('E_x - mean_x = ',  st.binom.mean(n,p) - np.mean(y))
```

```python
stat('binom', [1000,0.3])
```
